```python
import csv
import requests
from bs4 import BeautifulSoup

# Wikipedia URL with the list of literary works from Cesar Aira
url = 'https://es.wikipedia.org/wiki/C%C3%A9sar_Aira'

# Make the HTTP request and get the HTML content
response = requests.get(url)

```

```python
if response.status_code == 200:
    # Get the HTML content of the page
    html = response.content

    # Create a BeautifulSoup object to parse HTML content
    soup = BeautifulSoup(html, 'html.parser')

    # Obtain the text containing the works and years
    text_with_works = soup.get_text()

    # Divide the text into lines to analyze each one separately
    lines = text_with_works.split('\n')

    # Initialize a list to store the titles of novels and their years
    novels_and_years = []

```

```python
for line in lines:
    # Divide the line by a colon to separate the year from the title of the novel
    parts = line.split(':')
    if len(parts) >= 2:
        year = parts[0].strip()
        novel = ':'.join(parts[1:]).strip()
        novels_and_years.append({'Año': year, 'Título': novel})


```

```python
# Write data to a CSV file
with open('obras_y_años.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Año', 'Título']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for novel_year in novels_and_years:
        writer.writerow(novel_year)

print("Los datos se han guardado en 'obras_y_años.csv'")

# Donload the CSV file
from google.colab import files
files.download('obras_y_años.csv')

```
